<h2>그래프 그리기<h2>

```python
myundex05 = ['아메리카노', '에스프레소', '카페라떼', '카페모카', '콜드브루', '바닐라라떼', '마키아또']
mylist05 = [5000, 3000, 2000, 6000, 4000, 8000, 7000 ]
myseries05 = pd.Series(data=mylist05, index=myundex05)
myseries05
```

```python
myseries05.plot()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
myseries05.plot(kind='line', grid=True, rot=20, )
plt.xlabel('상품 명', fontsize=14)
plt.ylabel('단가', fontsize=14)
plt.title('품목별 단가', fontsize=18)
```

```python
myseries05.plot(kind='bar', grid=True, rot=0, color=['r','g','b','y','m'])
plt.xlabel('상품 명', fontsize=14)
plt.ylabel('단가', fontsize=14)
plt.title('품목별 단가', fontsize=18)
```
