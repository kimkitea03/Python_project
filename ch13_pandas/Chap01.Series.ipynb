<h6><font color="BlueViolet">시리즈의 값 읽어 오기</font></h6>

```python
myindex02 = ['아메리카노', '에스프레소', '카페라떼', '카페모카', '콜드브루', '바닐라라떼', '마끼아또']
mylist02 = [50, 30, 20, 60, 40, 80, 70]
myseries02 = pd.Series(data=mylist02, index=myindex02)
myseries02.to_frame()
```

```python
myseries02['카페모카'] # scalar로 반환
```

```python
# 콜론 기호는 연속된 데이터 추출
myseries02['에스프레소':'카페모카'] # 문자열 라벨링 : from <= 범위 <= to
```

```python
myseries02[['마끼아또', '카페모카']] # 콤마 기호는 서로 떨어진 데이터 추출
```

```python
myseries02.iloc[3] # i  loc
```

```python
myseries02.iloc[0:7:3]
```

```python
myseries02.iloc[3:6]
```

<h6><font color="BlueViolet">시리즈에 값 쓰기</font></h6>

```python
myseries02.iloc[2] = 2000
myseries02[2:5] = 5000
myseries02[['아메리카노', '카페모카']] = 3000
print(myseries02)
```

```python
myseries02[0::2] = 7000
print(myseries02)
```

```python
result = myseries02[myseries02 <= 4000]
print(result)
```

<h3><font color="red">시리즈 연산</font></h3>

<h6><font color="BlueViolet">시리즈 샘플 만들기</font></h6>

```python
myindex31 = ['에스프레소', '아메리카노', '마키아또', '바닐라라떼']
mylist31 = [30, 40, 50, 60]

myindex32 = ['에스프레소', '아메리카노', '마키아또', '콜드브루']
mylist32 = [20, 40, 60, 70]

myseries31 = pd.Series(data=mylist31, index=myindex31)
myseries32 = pd.Series(data=mylist32, index=myindex32)
```

```python
myseries31
```

```python
myseries32
```

```python
su = 10
print(myseries31 + su)
```

```python
print(myseries31.add(su))
```

```python
print(myseries31.sub(su))
```

```python
print(myseries31.mul(su))
```

```python
print(myseries31.div(su)) # floordiv() 함수도 있음
```

<h6><font color="BlueViolet">관계 연산자와 관련된 메소드</font></h6>

```python
myseries31[myseries31 >= 50]
```

```python
myseries31[myseries31.ge(50)]
```

```python
myseries31[myseries31.eq(50)]
```

<h6><font color="BlueViolet">시리즈간의 산술 연산</font></h6>

```python
newseries = myseries31 + myseries32 # NaN : Not a Number
print(newseries)
```

```python
newseries = myseries31.sub(myseries32, fill_value=0)
print(newseries)
```

```python
newseries = myseries31.mul(myseries32, fill_value=12)
print(newseries)
```

<h3><font color="red">유일값과 빈도 파악, 색인 필터링</font></h3>

<h6><font color="BlueViolet">unique 메소드 사용</font></h6>

```python
mylist04 = ['카페모카', '에스프레소', '에스프레소', '카페라떼', '아메리카노', '에스프레소', '아메리카노', '에스프레소', '카페모카', '카페모카']
myseries04 = pd.Series(mylist04)
myseries04
```

```python
myunique04 = myseries04.unique()
myunique04
```

```python
myseries04.value_counts() # 빈도수 파악해서 역순으로 정렬
```

```python
myseries04.value_counts(normalize=True) # 빈도수를 비율로 보여주기
```

```python
mask04 = myseries04.isin(['카페모카', '카페라떼'])
mask04
```

```python
myseries04[mask04]
```

<h3><font color="red">그래프 그리기</font></h3>

```python
myindex05 = ['아메리카노', '에스프레소', '카페라떼', '카페모카', '콜드브루', '바닐라라떼', '마끼아또']
mylist05 = [5000, 3000, 2000, 6000, 4000, 8000, 7000]
myseries05 = pd.Series(data=mylist05, index=myindex05)
myseries05
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
myseries05.plot(kind='line', grid=True, rot=20)
plt.xlabel('상품명', fontsize=14)
plt.ylabel('단가', fontsize=14)
plt.title('품목별 단가 꺽은 선', fontsize=18)
```

```python
myseries05.plot(kind='bar', grid=True, rot=0, color=['r', 'g', 'b', 'y', 'm'])
plt.xlabel('상품명', fontsize=14)
plt.ylabel('단가', fontsize=14)
plt.title('품목별 단가 꺽은 선', fontsize=18)
```
